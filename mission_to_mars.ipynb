```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://mars.nasa.gov/news/8744/nasa-engineers-checking-insights-weather-sensors/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
#Find title
news_title = soup.find('h1', class_="article_title")
print(news_title)
```

```python
#Find Paragraph Text
news_p = soup.find('div', class_="wysiwyg_content")
print(news_p)
```
